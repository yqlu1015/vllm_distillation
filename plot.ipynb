```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
file_path = "train_loss.csv"  # Replace with the path to your CSV file
data = pd.read_csv(file_path)

print(plt.style.available)
# Plot the loss curve
plt.style.use('seaborn-v0_8')
plt.figure(figsize=(8, 6), dpi=300)  # High resolution: 300 DPI
plt.plot(data['Step'], data['distillation-500 - loss/train'], color='blue', linewidth=1.5, label='Training Loss')

# Add labels, title, and grid
plt.xlabel("Step", fontsize=12)
plt.ylabel("Loss", fontsize=12)
# plt.title("Training Loss Curve", fontsize=14)
plt.grid(True)
# plt.legend()

# Save the high-resolution plot
plt.savefig("training_loss_curve.png", dpi=600, bbox_inches='tight')  # Save as PNG
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
topics = ['GEO', 'PHY', 'BIO', 'ERS', 'SAEP', 'CHE', 'USH']
num_training_data = [277, 317, 398, 93, 93, 122, 49]  # Histogram data
num_test_data = [600, 425, 403, 156, 128, 94, 84]
test_accuracies = [0.5783, 0.5624, 0.5385, 0.4615, 0.5391, 0.4574, 0.4048]  # Line chart data

# Number of topics
x = np.arange(len(topics))

# Create the figure and axes
fig, ax1 = plt.subplots(figsize=(10, 6))

# Bar chart: Number of training and test data
bar_width = 0.35
ax1.bar(x - bar_width/2, num_training_data, width=bar_width, color='skyblue', label='Training Data', alpha=0.7)
ax1.bar(x + bar_width/2, num_test_data, width=bar_width, color='lightgreen', label='Test Data', alpha=0.7)
ax1.set_xlabel("Topics", fontsize=12)
ax1.set_ylabel("Number of Data", fontsize=12, color='blue')
ax1.set_xticks(x)
ax1.set_xticklabels(topics, fontsize=10)
ax1.tick_params(axis='y', labelcolor='blue')

# Line chart: Test accuracies
ax2 = ax1.twinx()  # Create a second y-axis sharing the same x-axis
ax2.plot(x, test_accuracies, color='orange', marker='o', linewidth=2, label='Test Accuracy')
ax2.set_ylabel("Test Accuracy", fontsize=12, color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

# Add grid and title
ax1.grid(axis='y', linestyle='--', alpha=0.7)
# plt.title("Training/Test Data and Test Accuracy by Topic", fontsize=14)

# Add legends
ax1.legend(loc='upper right')
# ax2.legend(loc='upper right')

# Save and show the plot
plt.tight_layout()
plt.savefig("histogram_linechart_combined_with_two_columns.png", dpi=600)  # Save as high-res image
plt.show()
```
